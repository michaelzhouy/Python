## pivot() 需传入三个参数，第一个参数是用作行索引的列，第二个参数是用作列索引的列，第三个参数为数值列，用于填充DataFrame；遗漏最后一个参数，返回一个含有多层列的DataFrame

```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
```

```python
# 读取数据集
data = pd.read_csv('examples/macrodata.csv')
# 前5行
data.head()
```

```python
periods = pd.PeriodIndex(year=data.year,
                         quarter=data.quarter,
                         name='date')
```

```python
columns = pd.Index(['realgdp', 'infl', 'unemp'], name='item')
```

```python
data = data.reindex(columns=columns) # 只选这几列
```

```python
data.index = periods.to_timestamp('D', 'end')
```

```python
# stack(): 列→列
# reset_index() 行→列，并重新生成index
# rename() 重命名
ldata = data.stack().reset_index().rename(columns={0: 'value'})
```

```python
# 前10行
ldata[:10]
```

```python
# pivot() 传入三个参数，第一个参数是用作行索引的列，第二个参数是用作列索引的列，第三个参数为数值列，用于填充DataFrame
pivoted = ldata.pivot('date', # 用作行索引的列
                      'item', # 用作列索引的列
                      'value') # 用于填充DataFrame的数值列
```

```python
pivoted
```

```python
# 增加一个数值列
ldata['value2'] = np.random.randn(len(ldata))
```

```python
# 前10行
ldata[:10]
```

```python
# 遗漏最后一个参数，返回一个含有多层列的DataFrame
pivoted = ldata.pivot('date', # 用作行索引的列
                      'item') # 用作列索引的列
```

```python
# 前5行
pivoted[:5]
```

```python
# value列的前5行
pivoted['value'][:5]
```

```python
# set_index() 将列索引转换为行索引
# unstack() 行→列
unstacked = ldata.set_index(['date', 'item']).unstack('item')
```

```python
unstacked[:5]
```

## melt() 将多列合并成一列

```python
df = DataFrame({'key': ['foo', 'bar', 'baz'],
                'A': [1, 2, 3],
                'B': [4, 5, 6], 
                'C': [7, 8, 9]})
```

```python
df
```

```python
melted = pd.melt(df, 
                 ['key']) # 'key'作为唯一的分组指标
```

```python
melted
```

```python
# pivot()需传入三个参数，前两个参数是分别用作行和列索引的列，第三个参数是用于填充DataFrame的数值列
reshaped = melted.pivot('key', # 行索引
                        'variable', # 行索引
                        'value') # 用于填充DataFrame的数值列
```

```python
reshaped
```

```python
reshaped.reset_index() # key行→列，并重新生成行索引
```

```python
df
```

```python
pd.melt(df, 
        id_vars=['key'], # 分组指标
        value_vars=['A', 'B']) # 值列
```

```python
pd.melt(df, 
        value_vars=['A', 'B', 'C']) # 值列
```

```python
pd.melt(df, 
        value_vars=['key', 'A', 'B']) # 值列
```
