```python
fig = plt.figure() # 新建一个图片
plt.plot(np.random.randn(30).cumsum(), 'ko--') # 黑色虚线
```

```python
fig = plt.figure() # 新建一个图片
plt.plot(np.random.randn(30).cumsum(), 
         color='k', # 颜色黑色
         linestyle='dashed', # 线型为虚线
         marker='o') # 用圆点标记
```

```python
data = np.random.randn(30).cumsum()
```

```python
fig = plt.figure() # 新建一个图片
plt.plot(data, 
         'k--', # 黑色虚线
         label='Default') # 标签
plt.plot(data, 
         'k-', # 黑色实线
         drawstyle='steps-post', # 线型
         label='steps-post') # 标签
```

```python
plt.legend(loc='best')
```

# 刻度、标签和图例

## 设置标题、轴刻度、刻度和刻度标签

```python
fig = plt.figure() # 新建一个图片
```

```python
ax = fig.add_subplot(1, 1, 1) # 添加一个子图
```

```python
ax.plot(np.random.randn(1000).cumsum())
```

```python
# set_xticks在数据范围内设定刻度的位置
ticks = ax.set_xticks([0, 250, 500, 750, 1000]) # 设置x轴刻度点
```

```python
# set_xticklabels为标签赋值
labels = ax.set_xticklabels(['one', 'two', 'three', 'four', 'five'], 
                            rotation=30, # 旋转30度
                            fontsize='small') # 字体大小
```

```python
# 标题
ax.set_title('My first matplotlib plot')
```

```python
# 给x轴一个名称
ax.set_xlabel('Stage')
```

## 添加图例

```python
# 新建一个图片，添加一个子图
fig = plt.figure(); ax = fig.add_subplot(1, 1, 1)
```

```python
ax.plot(np.random.randn(1000).cumsum(), 
        'k', # 黑色
        label='one') # 标签
```

```python
ax.plot(np.random.randn(1000).cumsum(), 
        'k--', # 黑色虚线
        label='two') # 标签
```

```python
ax.plot(np.random.randn(1000).cumsum(), 
        'k.', # 黑色点 
        label='three') # 标签
```

```python
ax.legend(loc='best')
```

# 注释和子图加工

```python
fig = plt.figure() # 新建一个图片
ax = fig.add_subplot(1, 1, 1) # 添加一个子图

rect = plt.Rectangle((0.2, 0.75), 0.4, 0.15, color='k', alpha=0.3)
circ = plt.Circle((0.7, 0.2), 0.15, color='b', alpha=0.3)
pgon = plt.Polygon([[0.15, 0.015], [0.35, 0.4], [0.2, 0.6]],
                   color='g', alpha=0.5)

ax.add_patch(rect)
ax.add_patch(circ)
ax.add_patch(pgon)
```
