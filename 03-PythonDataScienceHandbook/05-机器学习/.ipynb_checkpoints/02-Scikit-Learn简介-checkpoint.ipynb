# Scikit-Learn的评估器API

## API基础知识
1. 通过从Scikit-Learn中导入适当的评估器类，选择模型类
2. 用合适的数值对模型类进行实例化，配置模型超参数
3. 整理数据，获取特征矩阵$X$和目标数组$y$
4. 调用模型实例的fit()方法对数据进行拟合
5. 对新数据应用模型
---
- 在有监督学习模型中，通常使用predict()方法预测新数据的标签；
- 在无监督学习模型中，通常使用transform()或predict()方法转换或推断数据的性质。

## 有监督学习示例：简单线性回归

```python
X_iris = iris.drop('species', 
                   axis=1) # 删除列
X_iris.shape
```

```python
y_iris = iris['species']
y_iris.shape
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
rng = np.random.RandomState(42)
x = 10 * rng.rand(50)
y = 2 * x - 1 + rng.randn(50)
plt.scatter(x, y)
```

(1) 选择模型类

```python
from sklearn.linear_model import LinearRegression
```

(2) 选择模型超参数

```python
model = LinearRegression(fit_intercept=True) # 是否拟合直线的截距
model
```

(3) 将数据整理成特征**矩阵**和目标**数组**

```python
X = x[:, np.newaxis]
X.shape
```

(4) 用模型拟合数据 

```python
model.fit(X, y)
```

```python
# 斜率
model.coef_
```

```python
# 截距
model.intercept_
```

(5) 预测新数据的标签

```python
xfit = np.linspace(-1, 11)
xfit
```

```python
Xfit = xfit[:, np.newaxis]
print('xfit.shape: {}, Xfit.shape{}'.format(xfit.shape, Xfit.shape))
```

```python
yfit = model.predict(Xfit)
```

```python
plt.scatter(x, y) # 散点图
plt.plot(xfit, yfit) # 曲线图
```

## 有监督学习示例：鸢尾花数据分类

```python
from sklearn.model_selection import train_test_split
Xtrain, Xtest, ytrain, ytest = train_test_split(X_iris, y_iris, 
                                                random_state=1)
```

```python
from sklearn.naive_bayes import GaussianNB  # 1. 选择模型类
model = GaussianNB()                         # 2. 初始化模型
model.fit(Xtrain, ytrain)                    # 3. 用模型拟合数据
y_model = model.predict(Xtest)               # 4. 对新数据进行预测
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(ytest, y_model)
```

## 无监督学习示例：鸢尾花数据降维

```python
from sklearn.decomposition import PCA # 1. 选择模型类
model = PCA(n_components=2)            # 2. 设置超参数，初始化模型
model.fit(X_iris)                      # 3. 拟合数据，注意这里不用y变量
X_2D = model.transform(X_iris)         # 4. 将数据转换为二维
```

```python
iris['PCA1'] = X_2D[:, 0] # 新增列
iris['PCA2'] = X_2D[:, 1] # 新增列
# 回归图
sns.lmplot('PCA1', 'PCA2', 
           hue='species', # 分类标识
           data=iris, # 数据源
           fit_reg=False) # 是否画出拟合出的直线
```

```python
# 回归图
sns.lmplot('PCA1', 'PCA2', 
           hue='species', # 分类标识
           data=iris, # 数据源
           fit_reg=True) # 是否画出拟合出的直线
```

## 无监督学习示例：鸢尾花数据聚类

```python
from sklearn.mixture import GaussianMixture    # 1. 选择模型类
model = GaussianMixture(n_components=3,
                        covariance_type='full') # 2. 设置超参数，初始化模型
model.fit(X_iris)
y_gmm = model.predict(X_iris)
```

```python
iris['cluster'] = y_gmm
sns.lmplot('PCA1', 'PCA2',
           data=iris, # 数据源
           hue='species', # 分类标识
           col='cluster', # 分列图
           fit_reg=False) # 是否画出拟合出的直线
```

# 应用：手写数字探索

```python
from sklearn.datasets import load_digits
digits = load_digits()
digits.images.shape
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(10, 10, # 10 × 10
                         figsize=(8, 8), # 图片大小
                         subplot_kw={'xticks': [], 'yticks': []}, # 刻度设置
                         gridspec_kw=dict(hspace=0.1, wspace=0.1)) # 子图之间的间距设置

# flat：10·10→100
for i, ax in enumerate(axes.flat):
    ax.imshow(digits.images[i], 
              cmap='binary', # 颜色
              interpolation='nearest')
    # 文本
    ax.text(0.05, 0.05, # 文本放置位置
            str(digits.target[i]), # 
            transform=ax.transAxes, # 
            color='green') # 颜色
```

```python
X = digits.data
X.shape
```

```python
y = digits.target
y.shape
```

```python
from sklearn.manifold import Isomap
iso = Isomap(n_components=2)
iso.fit(digits.data)
data_projected = iso.transform(digits.data)
data_projected.shape
```

```python
plt.scatter(data_projected[:, 0], # 第一个轴
            data_projected[:, 1], # 第二个轴
            c=digits.target, # 颜色
            edgecolor='none',
            alpha=0.5, # 透明度
            cmap=plt.cm.get_cmap('Spectral', 10)) # 颜色
# 颜色条
plt.colorbar(label='digit label', # 颜色条的label
             ticks=range(10)) # 刻度
plt.clim(-0.5, 9.5) # 范围设置
```

```python
Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, random_state=0)
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(Xtrain, ytrain)
y_model = model.predict(Xtest)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(ytest, y_model)
```

```python
from sklearn.metrics import confusion_matrix

mat = confusion_matrix(ytest, y_model) # 混淆矩阵
print(mat)

sns.heatmap(mat, square=True, annot=True, cbar=False)
plt.xlabel('predicted value')
plt.ylabel('true value')
```

```python
fig, axes = plt.subplots(10, 10, # 10×10个子图
                         figsize=(8, 8), # 图片大小
                         subplot_kw={'xticks': [], 'yticks': []}, # 刻度设置
                         gridspec_kw=dict(hspace=0.1, wspace=0.1)) # 子图之间的间距设置

test_images = Xtest.reshape(-1, 8, 8)

for i, ax in enumerate(axes.flat):
    ax.imshow(test_images[i], 
              cmap='binary', 
              interpolation='nearest')
    # 文本
    ax.text(0.05, 0.05,
            str(y_model[i]),
            transform=ax.transAxes,
            color='green' if (ytest[i] == y_model[i]) else 'red') # 误分类标识为红色
```
